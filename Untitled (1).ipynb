```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# KDE plot for income distribution
plt.figure(figsize=(8, 6))
sns.kdeplot(data=data, x='Income', fill=True, color='skyblue')
plt.xlabel('Income')
plt.ylabel('Density')
plt.title('Income Distribution')
plt.show()
```

```python
# KDE plot for income distribution
plt.figure(figsize=(8, 6))
sns.kdeplot(data=data, x='Recency', fill=True, color='skyblue')
plt.xlabel('Recency')
plt.ylabel('Density')
plt.title('Recency Distribution')
plt.show()
```

```python
# Box plot of amount spent on wines and gold products
plt.figure(figsize=(8, 6))
sns.boxplot(data=data[['MntWines', 'MntFruits','MntSweetProducts','MntGoldProds']])
plt.xlabel('Product Category')
plt.ylabel('Amount Spent')
plt.title('Amount Spent on Wines, Gold Products, Fruits and sweets')
plt.show()
```

```python
# Scatter plot of number of web visits and response
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='NumWebVisitsMonth', y='Response', hue='Response')
plt.xlabel('Number of Web Visits')
plt.ylabel('Response')
plt.title('Web Visits vs. Response')
plt.show()
```

```python
# Box plot of amount spent on wines and gold products
plt.figure(figsize=(8, 6))
sns.barplot(data=data[['NumWebPurchases', 'NumCatalogPurchases','NumStorePurchases']])
plt.xlabel('Number of Purchase')
plt.ylabel('Number')
plt.title('Purchasing Number of Catlog, Store, Web')
plt.show()
```

```python
# Select the categorical columns of interest
categorical_columns = ['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1', 'AcceptedCmp2', 'Complain']

# Count analysis for each categorical column
for column in categorical_columns:
    plt.figure(figsize=(8, 6))
    sns.countplot(data=data, x=column)
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.title(f'Count Analysis - {column}')
    plt.show()
```

```python
# KDE plot for income distribution
plt.figure(figsize=(8, 6))
sns.kdeplot(data=data, x='Age', fill=True, color='skyblue')
plt.xlabel('Recency')
plt.ylabel('Density')
plt.title('Recency Distribution')
plt.show()
```

```python
# Select the categorical columns of interest
categorical_columns = ['marital_Divorced', 'marital_Married', 'marital_Single', 'marital_Together', 'marital_Widow']

# Count analysis for each categorical column
for column in categorical_columns:
    plt.figure(figsize=(8, 6))
    sns.countplot(data=data, x=column)
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.title(f'Count Analysis - {column}')
    plt.show()
```

```python
# Analysis of MntTotal
plt.figure(figsize=(8, 6))
sns.histplot(data=data, x='MntTotal', bins=10, kde=True)
plt.xlabel('Total Amount Spent')
plt.ylabel('Count')
plt.title('Distribution of Total Amount Spent')
plt.show()
```

```python
# Analysis of MntRegularProds
plt.figure(figsize=(8, 6))
sns.histplot(data=data, x='MntRegularProds', bins=10, kde=True)
plt.xlabel('Amount Spent on Regular Products')
plt.ylabel('Count')
plt.title('Distribution of Amount Spent on Regular Products')
plt.show()
```

```python
# Analysis of AcceptedCmpOverall
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='AcceptedCmpOverall')
plt.xlabel('Campaign Overall Acceptance')
plt.ylabel('Count')
plt.title('Campaign Overall Acceptance Analysis')
plt.show()
```
